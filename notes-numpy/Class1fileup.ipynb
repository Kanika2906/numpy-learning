# Numpy Arrays

NumPy arrays are the main way we will use Numpy throughout the course. Numpy arrays essentially come in two flavors: vectors and matrices. Vectors are strictly 1-d arrays and matrices are 2-d (but you should note a matrix can still have only one row or one column).

Let's begin our introduction by exploring how to create NumPy arrays.

## Creating NumPy Arrays

### From a Python List

We can create an array by directly converting a list or list of lists:

```python
a = [1,2,3,4,5]
type(a)
```

```python
arr = np.array(a)
arr
```

```python
#in list we can add different datatypes , strings, numeric data or float data
#but in array if we use one element as string it will convert it to unicodes and will convert every element to strings
b = [1,"2",3,4,5]
barr = np.array(b)
barr.dtype
barr
```

Lets create a matrix using list

```python
#matrix means creating something in 2 dimensional space
a = [[1,2,3],[4,5,6],[7,8,9]]
arr = np.array(a)
arr
#if we will print a all elements will be printed in straight line but if we convert it into array then it will become 2 dimensional
```

Now lets see how matlix looks like using numpy arrays

```python
arr
```

## Built-in Methods

there are lots of build in methods to create an array lets create some and see

np.zeros(), np.ones(), np.full()

np.arange(), np.linspace()

np.eye()

### **np.zeros()**

```python
#you will create an array of zeros with np.zeros()
# one dimensional
arr = np.zeros(10)
arr
```

```python
# two dimensional
barr = np.zeros([5,6])
barr
```

**np.ones()**

```python
#similar working as np.zeros() the only change is the array created is of one
np.ones(5)
```

```python
np.ones([4,4])
```

**np.full()**

```python
np.full([5,5],7)
# 5 rows and 5 columns filled with number 7
```

**np.arrage()**

```python
#it works like range func , give starting and end point
np.arange(1,11)
```

**np.linspace()**

```python
# it arranges elements of a range in equal or linear space
#elements between 1 and 10 will be equally divided into 5
np.linspace(1,10,5)
```

**np.eye()**

```python
#eye is used to form identity matrix (1 in diagonal elements)
np.eye(5)
```

**np.random()**

```python
#will create an array with random values
np.random.rand(5,5)
#values will always be between 0 and 1
```

```python
#how to create random integers
#randint - random integer
#give range of number 1 to 10 and then size of array 3 (1D array) or [3,3] (2D array)
np.random.randint(1,10,[3,3])
```

```python
arr = np.random.randn(2,4)
#values between -3 and +3
arr
```

arange → step-based values

linspace → fixed number of values

eye → identity matrix

rand → random floats (0–1)

randint → random integers in a range

randn → random values from bell curve distribution
