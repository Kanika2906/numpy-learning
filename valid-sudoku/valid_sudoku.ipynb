Valid Sudoku

```python
import numpy as np
```

```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],

    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],

    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
```

```python
b = np.sum(s,axis=1)
for i in b:
  if i!=45:
    print("Sudoku is not valid.")
    break
else:
  print("Sudoku is valid for rows")
```

```python
b = np.sum(s,axis=0)
for i in b:
  if i!=45:
    print("Sudoku is not valid.")
else:
  print("Sudoku is valid for columns")
```

```python
for i in range(0,9,3):
  for j in range(0,9,3):
    if np.sum(s[i:i+3,j:j+3])!=45:
      print("Sudoku is not valid.")
else:
  print("Sudoku is valid")
```
